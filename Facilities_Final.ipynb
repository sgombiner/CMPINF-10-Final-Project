**First Lets Create a pie chart includes the different types of facilities**

```python
# import pandas
import pandas as pd
import matplotlib.pyplot as plt
# Load the file
filePath = 'fbb50b02-2879-47cd-abea-ae697ec05170.csv'  
data = pd.read_csv(filePath)
# Calculate the count of facility
typeCounts = data['type'].value_counts()
# Define a filter
Filter = 0.02 * len(data)
# Sum the counts of all types below 0.02 as 'Other'
other = typeCounts[typeCounts < Filter].sum()
typeCounts = typeCounts[typeCounts >= Filter]
typeCounts['Other'] = other
# Create a pie chart
plt.figure(figsize=(10, 6))
typeCounts.plot(kind='pie', autopct='%1.1f%%', startangle=90, ylabel='')
plt.title('Distribution of Facility Types in Pittsburgh')
plt.show()
```
