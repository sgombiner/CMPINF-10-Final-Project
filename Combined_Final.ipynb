# Introduction

For our project, we decided to determine the best neighborhood in Pittsburgh based on one metric: how much each neighborhood values outdoor entertainment. Though many factors are involved in what truly determines the "best" neighborhood to live in, how much a city and its neighborhoods value its avenues of recreation can serve as a good gauge to the wellbeing and quality of life of the areas in question. At first, we chose "entertainment" in general as a working metric, before changing it solely to "focus on outdoor entertainment" due to how much of our datasets leaned towards outdoors activities.

Our approach will involve taking three combinations of two datasets, calculating the percentages for both datasets, before adding them together for each neighborhood and sorting the combined percentages from greatest to lowest and plotting the top ten highest neighborhood in a bar chart. We will repeat this approach for all three possible double-combinations of our datasets, before combining all three together into a final measurement.

# The Metric

The metric is the **percentage of available avenues of outdoor recreation and entertainment in a Pittsburgh neighborhood.** To measure this metric, we used three different submetrics based off of different datasets: * The number of outdoor recreation facilities (excluding Parks and Baseball Fields) per neighborhood (i.e. recreational centers, pools, etc.)
* The number of parks per neighborhood
* The number of baseball fields per neighborhood

The datasets used were:

* [City of Pittsburgh Facilities](https://data.wprdc.org/dataset/city-of-pittsburgh-facilities/resource/fbb50b02-2879-47cd-abea-ae697ec05170) - City Facility data pulled from the Operations Management System for the Department of Public Works. Contains data for most, if not all known facilities in Pittsburgh
* [City of Pittsburgh Parks](https://data.wprdc.org/dataset/parks) - A collection of parks within Pittsburgh
* [Field Listings](https://data.wprdc.org/dataset/field-listings) - Contains data about the baseball fields within Pittsburgh

# The analysis

## First combination: Facilities and Parks

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first CSV file into a Pandas DataFrame
data1 = pd.read_csv('fbb50b02-2879-47cd-abea-ae697ec05170.csv')
neighborhood_counts1 = data1['neighborhood'].value_counts().sort_values(ascending=False)

# Read the second CSV file into a Pandas DataFrame
data2 = pd.read_csv('fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv')
neighborhood_counts2 = data2['neighborhood'].value_counts().sort_values(ascending=False)


# Calculate percentages for each dataset
total1 = neighborhood_counts1.sum()
percentages1 = (neighborhood_counts1 / total1) * 100

total2 = neighborhood_counts2.sum()
percentages2 = (neighborhood_counts2 / total2) * 100

# Combine the datasets, summing up the counts for each neighborhood
combined_neighborhood_counts = percentages1.add(percentages2, fill_value=0)

# Sort the combined percentages from greatest to lowest
combined_counts_sorted = combined_neighborhood_counts.sort_values(ascending=False)

# Selecting the top 10 neighborhoods
top_10_combined = combined_counts_sorted.head(10)

# Plotting Bar Chart of Number of Parks By Neighborhood
plt.figure(figsize=(12, 8))
top_10_combined.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Percentage of Facilities and Parks (%)')
plt.title('Combined Percentage of Facilities and Parks in Top 10 Neighborhoods')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
