```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first CSV file into a Pandas DataFrame
data1 = pd.read_csv('fbb50b02-2879-47cd-abea-ae697ec05170.csv')
neighborhood_counts1 = data1['neighborhood'].value_counts()

# Read the second CSV file into a Pandas DataFrame
data2 = pd.read_csv('fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv')
neighborhood_counts2 = data2['neighborhood'].value_counts()

# Calculate percentages for each dataset
total1 = neighborhood_counts1.sum()
percentages1 = neighborhood_counts1 / total1

total2 = neighborhood_counts2.sum()
percentages2 = neighborhood_counts2 / total2

# Combine the datasets, summing up the counts for each neighborhood
combined_neighborhood_counts = percentages1.add(percentages2, fill_value=0)

# Sort the combined percentages from greatest to lowest
combined_counts_sorted = combined_neighborhood_counts.sort_values(ascending=False)

# Selecting the top 10 neighborhoods
top_10_combined = combined_counts_sorted.head(10)

# Plotting Bar Chart of Number of Parks By Neighborhood
plt.figure(figsize=(12, 8))
top_10_combined.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Percentage of Facilities and Parks')
plt.title('Combined Percentage of Facilities and Parks in Top 10 Neighborhoods')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the first CSV file into a Pandas DataFrame
data1 = pd.read_csv('fbb50b02-2879-47cd-abea-ae697ec05170.csv')
neighborhood_counts1 = data1['neighborhood'].value_counts()

# Read the second CSV file into a Pandas DataFrame
data2 = pd.read_csv('fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv')
neighborhood_counts2 = data2['neighborhood'].value_counts()

# Read the third CSV file into a Pandas DataFrame
data3 = pd.read_csv('f9cdb902-d135-4c33-bc09-221f3a3c9c1f.csv')
neighborhood_counts3 = data3['Neighborhood'].value_counts() #case sensitive

# Combine the datasets, summing up the counts for each neighborhood
combined_counts = neighborhood_counts1.add(neighborhood_counts2, fill_value=0).add(neighborhood_counts3, fill_value=0)#fill value say's NA = 0

# Calculate percentages for the combined dataset
total = combined_counts.sum()
percentages = combined_counts / total

# Sort the combined percentages from greatest to lowest
combined_counts_sorted = percentages.sort_values(ascending=False)

# Selecting the top 10 neighborhoods
top_10_combined = combined_counts_sorted.head(10)

# Plotting Bar Chart of Number of Facilities and Parks By Neighborhood
plt.figure(figsize=(12, 8))
top_10_combined.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Percentage of Facilities, Parks and Baseball Fields')
plt.title('Combined Percentage of Facilities, Parks and Baseball Fields in Top 10 Neighborhoods (Sorted)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
