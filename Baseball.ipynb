# Premise

This notebook seeks to analyze the ["Field Listings" dataset](https://data.wprdc.org/dataset/field-listings) from the Western Pennsylvania Regional Data Center, with the main goal being **determining which neighborhood in Pittsburgh** has the most baseball fields.

The reason for choosing baseball fields as a submetric is to gauge how many resources Pittsburgh's governing body allocates towards recreational avenues for each neighborhood.

The basic procedure of gathering and analyzing the datasets follows as such:

1. Import the dataset into a pandas DataFrame via the CSV file
2. Using the "neighborhood" column of the CSV, count how many baseball fields there are in each neighborhood (done through value_counts)
3. Make a top ten
4. For the first block, make a bar plot of EVERY neighborhood with their baseball field count
5. For the second block, make a bar plot of the TOP TEN neighborhoods
6. For the third plot, make a PIE plot of the TOP TEN neighborhoods

```python
#import the necessary libraries

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import math

# Set style
plt.style.use('seaborn-notebook')

# Import the CSV file into a pandas DataFrame
baseball = pd.read_csv("baseball.csv", sep=",")

# Count up how many baseball fields each neighborhood has and then sort by descending value
neighborhoods = baseball["neighborhood"].value_counts().sort_values(ascending=False)

# Print the abbreviated list of neighborhoods
print(neighborhoods)

# make a bar plot of EVERY neighborhood with their baseball field count
neighborhoods.plot.bar(figsize=(12, 8), ylabel="Number of baseball fields")
plt.tight_layout()
plt.title('Distribution of Baseball Fields in Pittsburgh by Neighborhood')
```

As we can see, the bar plot goes through an exponential decay, showing that the range of the number of baseball fields per neighborhood is wide from Highland Park's count to Middle Hill's count.

Since we have a lot of neighborhoods to work with, let's narrow down the selection down to the top 10 neighborhoods.

```python
# get the top 10 neighborhoods with the most baseball fields
topTen = neighborhoods.head(10)

# print the top 10 neighborhoods
print(topTen)

# Make a bar chart of the top 10 neighborhoods with the most baseball fields
topTen.plot.bar(figsize=(12, 8), ylabel="Number of baseball fields")
plt.tight_layout()


# Make a pie chart of the top 10 neighborhoods with the most baseball fields, using autopct to display their distribution amongst the top 10 in percent form
plt.figure(figsize=(10, 6))
topTen.plot.pie(autopct='%1.1f%%', ylabel=None)
plt.title('Distribution of Baseball Fields in Pittsburgh by Neighborhood (Top 10)')
plt.show()
```

# The Answer

From the data gathered and sorted, Highland Park is the neighborhood with the greatest number of baseball fields within the city of Pittsburgh. However, Squirrel Hill South comes very close with 29 baseball fields within its vicinity. Both neighborhoods are very close to each other in terms of baseball field counts, but there is a significant gap in the number between Squirrel Hill South and the 3rd place, Perry North. 17 neighborhoods are tied for the lowest number of baseball fields in Pittsburgh (1), showing how centralized the presence of baseball fields are in the city.
