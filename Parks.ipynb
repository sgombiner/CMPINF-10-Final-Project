**Introduction**

* Dataset: City Parks in Pittsburgh
* Reasoning: Parks are integral to a social and flourishing neighborhood, not only as a place of greenery but as a common social space. Parks are also a generally good metric for determining a neirborhoods outdoors entertainment value, which is our metric.
* Code Discription: 
    * The first code bit sorts all of the parks in my dataset using the neighborhood column by frequency, summing each neighborhoods park total. It then plots a percentage pie chart and a quantitative bar chart.
    * The second code bit simplifies the bar and pie chart by the top 10 neighborhoods by park amount, to allow for more digestable viewing. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
parks_df = pd.read_csv('fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv')
neighborhood_counts = parks_df['neighborhood'].value_counts()

#Plotting Pie Chart of Number of Parks By Neighborhood
plt.figure(figsize=(10, 8))
plt.pie(neighborhood_counts, labels=neighborhood_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Parks Across Neighborhoods')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


#Plotting Bar Chart of Number of Parks By Neighborhood
plt.figure(figsize=(12, 8))
neighborhood_counts.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.title('Number of Parks in Each Neighborhood')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

**Analysis:**

My code sorted the Parks Dataset using the column of neighborhood to represent the frequency of parks per neighborhood in Pittsburgh, and then displayed the percentages in a pie chart and the amounts in a bar graph. East Liberty takes up a surprising 5.7% of total parks in Pittsburgh with 12 total parks. Central Buisness Disctrict follows closely behind with 4.8% of total parks in Pittsburgh with 10 parks. Third was Beechview, fourth was South Side Slopes, and fifth was Point Breeze. You'll notice that while Squirrel Hill South only comes in 6th place in this dataset, it still won our vote for best Neighborhood. This is because of it's high number of not only parks, but also facilities and baseball fields.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a Pandas DataFrame
parks_df = pd.read_csv('fa329e3d-89ff-4708-8ddf-81bfedcad11d.csv')
neighborhood_counts = parks_df['neighborhood'].value_counts()

# Selecting the top 10 neighborhoods
top_10_neighborhoods = neighborhood_counts.head(10)

# Plotting Pie Chart of Number of Parks By Neighborhood
plt.figure(figsize=(10, 8))
plt.pie(top_10_neighborhoods, labels=top_10_neighborhoods.index, autopct='%1.1f%%')
plt.title('Distribution of Parks Across Top 10 Neighborhoods')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Plotting Bar Chart of Number of Parks By Neighborhood
plt.figure(figsize=(12, 8))
top_10_neighborhoods.plot(kind='bar')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Parks')
plt.title('Number of Parks in Top 10 Neighborhoods')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
