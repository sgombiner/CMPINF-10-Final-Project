```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'fbb50b02-2879-47cd-abea-ae697ec05170.csv'  
data = pd.read_csv(file_path)

type_counts = data['type'].value_counts()

threshold = 0.02 * len(data)
other = type_counts[type_counts < threshold].sum()
type_counts = type_counts[type_counts >= threshold]
type_counts['Other'] = other

plt.figure(figsize=(10, 6))
type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, ylabel='')
plt.title('Distribution of Facility Types in Pittsburgh')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'fbb50b02-2879-47cd-abea-ae697ec05170.csv'  
data = pd.read_csv(file_path)

neighborhood_counts = data['neighborhood'].value_counts()

threshold = 0.02 * len(data)
other = neighborhood_counts[neighborhood_counts < threshold].sum()
neighborhood_counts = neighborhood_counts[neighborhood_counts >= threshold]
neighborhood_counts['Other'] = other

plt.figure(figsize=(10, 6))
neighborhood_counts.plot(kind='bar', ylabel='Number of Facilities')
plt.title('Distribution of Facilities by Neighborhood in Pittsburgh')
plt.xticks(rotation=45)
plt.show()
```
